# Question 6

```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_csv('us-marriages-divorces-1867-2014.csv')
data.head()
plt.figure( figsize=(10, 8))
data.plot(x = 'Year', y = ['Marriages_per_1000','Divorces_per_1000'],
label = ['Marraiges' , 'Divorces'])
plt.title('Marraiges & Divorces Count Per Capita', fontsize = 18)
plt.xlabel('Year', fontsize = 14)
plt.ylabel('Count per Capita', fontsize = 14)
plt.show()
```

# Question 7

```python
Data_BarChart=data[data['Year'].isin([1900,1950,2000])][['Year','Marriages_per_1000','Divorces_per_1000']]
```

```python
import matplotlib.pyplot as plt
Data_BarChart[['Marriages_per_1000','Divorces_per_1000']].plot.bar(figsize=(10,8))
plt.title("Marriages and Divorces Per 1000", fontsize = 18)
plt.xlabel('Year', fontsize = 14)
plt.ylabel('Count per Capita', fontsize = 14)
positions = range(len(Data_BarChart))
labels = Data_BarChart['Year']
plt.xticks(positions, labels, rotation = 0)
plt.show()
```
